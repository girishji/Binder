Price is proportional to marginal utility. Treat y axis as marginal utility w.r.t. water quantity $$\frac{\delta U}{\delta q}$$

```python
%matplotlib inline
import sympy as sym
from sympy.plotting import plot

q, p_a, p_b, U_a, U_b = sym.symbols('q p_a p_b U_a U_b')

MNB_A = -p_a - 1.25*q + 125
Price_A = sym.solveset(MNB_A, p_a).args[0]
pl = plot(Price_A, (q, 0, 150), ylim=[0, 300], xlabel='quantity', ylabel='price', show=False, line_color='red', \
          title='A=red, B=blue')
MNB_B = -p_b - (150/450)*q + 150
Price_B = sym.solveset(MNB_B, p_b).args[0]
pl.extend(plot(Price_B, (q, 0, 500), show=False))
pl.show()
```

Utility function should be of the form aX^2 + bX.

```python
Utility_A = -U_a - (1.25/2)*q**2 + 125*q
Utility_B = -U_b - (150/900)*q**2 + 159*q

Util_A = sym.solveset(Utility_A, U_a).args[0]
pl = plot(Util_A, (q, 0, 200), xlabel='Utility', ylabel='price', show=False, line_color='red', \
          title='A=red, B=blue')
Util_B = sym.solveset(Utility_B, U_b).args[0]
pl.extend(plot(Util_B, (q, 0, 800), show=False))
pl.show()
```

```python
p_d, q_d, p_s, q_s = sym.symbols('p_d q_d p_s q_s')

Eq_d = 1000 - 5*p_d - q_d
Eq_s = -250 + 2*p_s - q_s
Eq_eqm_q = q_d - q_s
Eq_eqm_p = p_d - p_s

# Equilibrium
solution_eqm = sym.solve((Eq_d, Eq_s, Eq_eqm_q, Eq_eqm_p), (q_d, q_s, p_d, p_s))
solution_eqm
```

```python
# demand curve
Price_d = sym.solveset(Eq_d, p_d).args[0]
pl = plot(Price_d, (q_d, 0, 1000), xlabel='quantity', ylabel='price', show=False, line_color='red', \
          title='demand=red, supply=blue')
# supply curve
Price_s = sym.solveset(Eq_s, p_s).args[0]
pl.extend(plot(Price_s, (q_s, 0, 1000), show=False))
pl.show()
```

```python
# External cost
EC = 10*q_d + 0.01*(q_d**2)

print(f'External cost: {EC.subs(q_d, solution_eqm[q_d]):.2f}')
```
